```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
print(df.head())
```

```python
plt.figure(figsize=(6, 6))
df['day'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange', 'pink'])
plt.title("Distribution of Days")
plt.ylabel('')  # Hide y-label for clarity
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['total_bill'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()


sns.histplot(df["tax_paid"], bins=20, kde=True)

plt.title("Distribution of Tax paid")

plt.show()
```

```python

sns.histplot(df["total_bill"], bins=20, kde=True)

plt.title("Distribution Bills")

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="day", y="total_bill", data=df, estimator=sum, palette="magma")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.title("Total Bill per Day")
plt.show()
```
